```python
import numpy as np
import matplotlib.pyplot as plt
```

### Train

```python
all_samples = np.load('/home/project/code/data/scannet_splits/smart_sample_train.npz')
```

```python
pairs = all_samples['name']
scores = all_samples['score']
N = pairs.shape[0]
```

```python
np.random.seed(10)
subset = np.random.choice(N, 10_000, replace=False)
```

```python
plt.hist(scores[subset], bins=20);
```

```python
np.savez('/home/project/code/data/scannet_splits/train_subset_ablations.npz', name=pairs[subset, :], score=scores[subset])
```
