```python
fig = plt.figure(constrained_layout=True, figsize=(10, 5))
fig.suptitle('Results on test subset of ScanNet')

subfig1, subfig2, subfig3 = fig.subfigures(3, 1)
ax00, ax01, ax02, ax03 = subfig1.subplots(1, 4)   

sns.ecdfplot(x=x, complementary=False, ax=ax00)
ax00.set_xlabel('Angle error', fontsize=8)
ax00.set_title('CDF of angle error', fontsize=8)
ax00.grid()


subfig1.suptitle('GMFlow + dense regressor w/o confidences', fontsize=10)
ax10, ax11, ax12, ax13 = subfig2.subplots(1, 4)  
subfig2.suptitle('GMFlow + confidence module + dense regressor', fontsize=10) 

ax20, ax21, ax22, ax23 = subfig3.subplots(1, 4)  
subfig3.suptitle('GMFlow + confidence module + E estimator', fontsize=10); 
```

```python
x1 = np.random.rand(10000)
x2 = np.random.rand(10000) * 2
x3 = np.random.rand(10000) * 3 
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

title_fontsize=10
label_fontsize=8


sns.ecdfplot(x=x1, complementary=False, ax=ax[0,0], c='b')
sns.ecdfplot(x=x2, complementary=False, ax=ax[0,0], c='r')
sns.ecdfplot(x=x3, complementary=False, ax=ax[0,0], c='c')

ax[0, 0].set_xlabel('Angle error', fontsize=title_fontsize)
ax[0, 0].set_title('CDF of angle error', fontsize=label_fontsize)
ax[0, 0].grid()


sns.ecdfplot(x=x1, complementary=False, ax=ax[0,1])
ax[0, 1].set_xlabel('$\|t_{gt} - t\|_1$')
ax[0, 1].set_title('CDF of translation error')
ax[0, 1].grid()


sns.ecdfplot(x=x1, complementary=False, ax=ax[1,1], c='b', label='GMFlow + dense regressor w/o confidences')
sns.ecdfplot(x=x2, complementary=False, ax=ax[1,1], c='r', label='GMFlow + confidence module + dense regressor')
sns.ecdfplot(x=x3, complementary=False, ax=ax[1,1], c='c', label='GMFlow + confidence module + E estimator')

ax[1, 1].set_xlabel('Angle error', fontsize=title_fontsize)
ax[1, 1].set_title('CDF of angle error', fontsize=label_fontsize)
ax[1, 1].grid()
handles, labels = ax[1, 1].get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center')
# plt.tight_layout()


# sns.ecdfplot(x=x1, complementary=False, ax=ax[0,1])

# ax[0, 1].set_xlabel('$\|t_{gt} - t\|_1$')
# ax[0, 1].set_title('CDF of translation error')
# ax[0, 1].grid()
```
