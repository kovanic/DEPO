```python
np.random.seed(10)

x = np.random.laplace(loc=2, scale=10, size=100_000)
y = np.random.laplace(loc=10, scale=10, size=100_000)
z = np.random.laplace(loc=20, scale=10, size=100_000)

n = x**2 + y**2 + z**2

x_n = x / n
y_n = y / n
z_n = z / n

fig, ax = plt.subplots(2, 3, figsize=(10, 5))
sns.histplot(x, bins=50, ax=ax[0][0])
sns.histplot(y, bins=50, ax=ax[0][1])
sns.histplot(z, bins=50, ax=ax[0][2])

sns.histplot(x_n, bins=50, ax=ax[1][0])
sns.histplot(y_n, bins=50, ax=ax[1][1])
sns.histplot(z_n, bins=50, ax=ax[1][2])
```
